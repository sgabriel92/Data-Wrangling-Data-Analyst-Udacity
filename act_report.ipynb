## Report: WeRateDogs Insights
### Introduction
The tiwtter account WeRateDogs rates people's dogs with a humorous comment about the dog. These ratings
almost always have a denominator of 10. The numerators, though? Almost always greater than 10. 11/10, 12/10, 13/10, etc. WeRateDogs has over 4 million followers and has received international media coverage. 
Twitters API to access this data, combined with a dataset from a neural network that predicted various dog breeds and a twitter archive file gives a great baseline to gather, assess and clean the data.The final product after the wrangling process allows analysis and visualization of insights retrieved from the data.

### Insights

The cleaned and combined data allows insights invarious topics. One of them is the most common dog name that was used.
